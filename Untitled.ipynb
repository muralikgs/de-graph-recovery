```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
nl = 75
nh = 25
```

```python
A = np.load("../pref_req_data/sensing_matrix_de.npy")
A_bs = np.load("../pref_req_data/sensing_matrix_bs.npy")
```

```python
int_num = 83

Cov_X = np.load("../pref_req_data/ground_truth/cov_x_{}.npy".format(int_num))

indices_list = [
    np.load("../pref_req_data/de_measurements/inter_{}/ind_set_{}.npy".format(int_num, t)) for t in range(5)
]

Cov_X_rec = [
    np.load("../pref_req_data/de_recovered/inter_{}/X_cov_rec_{}.npy".format(int_num, t)) for t in range(5)
]

Cov_X_rec_bs = [
    np.load("../pref_req_data/bs_recovered/inter_{}/X_cov_rec_{}.npy".format(int_num, t)) for t in range(5)
]
```

```python
for i in range(2):
    col_ind_r, row_ind_r = np.meshgrid(indices_list[i], indices_list[i])

    Cov_X_r = Cov_X[row_ind_r, col_ind_r]

    Cov_X_de_r = Cov_X_rec[i][row_ind_r, col_ind_r]
    Cov_X_bs_r = Cov_X_rec_bs[i][row_ind_r, col_ind_r]

    Cov_X_hh = Cov_X_r[:nh, :nh]
    Cov_x_hat_de_hh = Cov_X_de_r[:nh, :nh]
    Cov_x_hat_bs_hh = Cov_X_bs_r[:nh, :nh]

    print("DE Error, Fro: {}, MAE: {}".format(np.linalg.norm(Cov_X_hh - Cov_x_hat_de_hh, 'fro'), np.abs(Cov_X_hh - Cov_x_hat_de_hh).max()))
    print("BS Error, Fro: {}, MAE: {}".format(np.linalg.norm(Cov_X_hh - Cov_x_hat_bs_hh, 'fro'), np.abs(Cov_X_hh - Cov_x_hat_bs_hh).max()))
    print()
    
```

```python
%matplotlib notebook

fig, axs = plt.subplots(1, 3)

axs[0].imshow(Cov_X_hh)
axs[1].imshow(Cov_x_hat_de_hh)
axs[2].imshow(Cov_x_hat_de_hh)
```
